```python
# which team won the most matches
match_win = df['match_winner'].value_counts()
match_win
```

```python
sns.barplot(y = match_win.index, x = match_win.values, palette = 'viridis')
plt.title("Most Matches Won By Teams", fontsize = 12)
```

```python
# toss decision trens
df['toss_decision'].value_counts()
sns.countplot(x = df['toss_decision'], palette = 'mako')
plt.title("Toss Decision Trends", fontsize = 10)
```

```python
# toss Winner VS Match Winner
count = df[df["toss_winner"] == df["match_winner"]]['match_id'].count()
percentage = (count*100)/df.shape[0]
percentage.round(2)
```

```python
# how do teams win matches (by runs or wickets)
df['won_by'].value_counts()
sns.countplot(x = df['won_by'], palette = 'cubehelix')
plt.title("Ways to Win Matches", fontsize = 10)
```

```python
#  player of the match winners
count = df['player_of_the_match'].value_counts().head(10)
sns.barplot(y = count.index, x = count.values, palette = 'rocket')
plt.title("Player of the Match ", fontsize = 12)
```

```python
df.head()
```

```python
# 2 top scorer 
hightest_score = df.groupby('top_scorer')['highscore'].sum().sort_values(ascending= False).head(2)
hightest_score 
```

```python
hightest_score.plot(kind = 'barh')
plt.title("Top Scorers in IPL", fontsize = 10)
```

```python
df.head()
```

```python
# 10 Best bowling Figures

df['highest_bowling'] = df['best_bowling_figure'].apply(lambda x :x.split('--')[0])
df['highest_bowling'] = df['highest_bowling'].astype(int)

top_bowler = df.groupby('best_bowling')['highest_bowling'].sum().sort_values(ascending = False).head(10)
top_bowler
```

```python
top_bowler.plot(kind= 'barh', color = 'green')
plt.title("Top 10 Best Bowling Figures", fontsize = 10)
```

```python
# most matches played by vanue
df.head()
df['venue'].value_counts().head(10)
sns.barplot(y = df['venue'].value_counts().head(10).index, x = df['venue'].value_counts().head(10).values, palette = 'plasma')
plt.title("Top 10 Venues Hosting Most Matches", fontsize = 10)
```
